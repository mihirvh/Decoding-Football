```python
## FUNCTION TO SEGREGATE NUMERICAL AND CATEGORICAL COLUMNS FOR EASY EDA
def data_type(df):
    num = df.select_dtypes(include=["int64", "float64"]).columns.tolist()
    cat = [x for x in df.columns if x not in num]
    return num, cat
    
numerical, categorical = data_type(games_df)



changes = ["home_club_position", "away_club_position"]
for i in changes:
    numerical.remove(i)
categorical.extend(changes)


print("numerical: ", numerical)
print()
print("categorical: ", categorical)
```

```python
## GETTING THE DISTRIBUTION FOR EACH NUMERICAL COLUMNS FOR MISSING VALUE IMPUTATION
def distribution(df):
    col = 2
    rows = len(numerical)//2 + 1
    fig, ax = plt.subplots(rows, 2, figsize=(20, 10))
    ax = ax.flatten()
    
    for i, column in enumerate(df[numerical].columns):
        sns.histplot(data=df[column], kde=True, ax=ax[i])
        ax[i].set_title(f'Distribution of {column}')
    
    # Adjust layout
    plt.tight_layout()
    plt.show()
distribution(games_df)
```

```python
def impute_missing_vals(df):
    for i in df.columns:
        if i in categorical:
            mode = df[i].mode()[0]
            df[i].fillna(mode, inplace = True)
        elif i in numerical:
            median = df[i].median()
            df[i].fillna(median, inplace = True)
                
impute_missing_vals(games_df)
```

### OUTLIER TREATMENT

```python
## COUNTING OUTLIERS IN EACH COLUMNS
def outlier_counts(df, threshold=1.5):
    outlier_count = {}


    for i in df[numerical].columns:
        # Calculate Q1 (25th percentile) and Q3 (75th percentile)
        q1 = df[i].quantile(0.25)
        q3 = df[i].quantile(0.75)

        # Calculate Interquartile Range (IQR)
        iqr = q3 - q1

        # Calculate lower and upper fences
        lf = q1 - (iqr * threshold)
        uf = q3 + (iqr * threshold)

        # Initialize the outlier count for this column
        outlier_count[i] = 0
        
        # Count the number of values below the lower fence and above the upper fence
        lower_outliers = (df[i] < lf).sum()
        upper_outliers = (df[i] > uf).sum()

        # Total outliers is the sum of lower and upper outliers
        outlier_count[i] = lower_outliers + upper_outliers

    return outlier_count
    
outlier_count = outlier_counts(games_df)

outlier_count
```

```python
## TREATING OUTLIERS IN EACH COLUMN 
def winsorization(df, threshold = 1.5):
    winsorized_df = df.copy()
    for i in winsorized_df[numerical].columns:
        q1 = winsorized_df[i].quantile(0.25)
        q3 = winsorized_df[i].quantile(0.75)
        
        iqr = q3 - q1

        lf = q1 - (iqr * threshold)
        uf = q3 + (iqr * threshold)
        

        if winsorized_df[i].dtype == "float64":
            winsorized_df.loc[winsorized_df[i] < lf, i] = lf
            winsorized_df.loc[winsorized_df[i] > uf, i] = uf
        elif winsorized_df[i].dtype == "int64":
            winsorized_df.loc[winsorized_df[i] < lf, i] = int(lf)
            winsorized_df.loc[winsorized_df[i] > uf, i] = int(uf)
    return winsorized_df   
games_df_winsorised = winsorization(games_df)

#players_df_winsorised
```

```python
## CHECKING FOR VALID OUTLIER TREATMENT
outlier_count = outlier_counts(games_df_winsorised)
outlier_count
```
